# Cropped - VGG16 x Original

## Accuraccy

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os arquivos CSV
vgg16_cropped = pd.read_csv('training_history_cropped_vgg16.csv')
original_cropped = pd.read_csv('training_history_cropped_original.csv')

# Criar o gráfico de linhas para 'accuracy' e 'val_accuracy'
plt.figure(figsize=(12, 6))

# Plotando os dados da VGG16 (cropped)
plt.plot(vgg16_cropped['accuracy'], label='VGG16 - Accuraccy (Treinamento)', linestyle='-', marker='o')
plt.plot(vgg16_cropped['val_accuracy'], label='VGG16 - Accuraccy (Validação)', linestyle='--', marker='o')

# Plotando os dados da Rede Original (cropped)
plt.plot(original_cropped['accuracy'], label='Personalizada - Accuraccy (Treinamento)', linestyle='-', marker='s')
plt.plot(original_cropped['val_accuracy'], label='Personalizada - Accuraccy (Validação)', linestyle='--', marker='s')

# Configurações do gráfico
plt.title('Comparação de Accuraccy - VGG16 vs. Rede Personalizada (Crooped)')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.grid(True)

# Exibir o gráfico
plt.show()
```

# Overfiting

```python
#Gráfico de Overfitting – Visualiza a diferença entre treino e validação para identificar overfitting.

def plot_overfitting(tabela, nome):
    plt.figure(figsize=(8, 5))
    plt.plot(tabela['loss'] - tabela['val_loss'], label='Diff (Train - Val) Loss', color='red')
    plt.title(f'Overfitting Analysis - {nome}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss Difference')
    plt.grid(True)
    plt.legend()
    plt.show()

# Exemplo para a Cropped Original
plot_overfitting(tabela1, 'Cropped Personalizada')
plot_overfitting(tabela2, 'Cropped VGG16')
plot_overfitting(tabela3, 'Normal Personalizada')
plot_overfitting(tabela4, 'Normal VGG16')
```
